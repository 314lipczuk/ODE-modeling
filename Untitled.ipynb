```python
def ramp_light_fn(t, rest):
    # Smooth transition half-width
    delta_t = 0.2

    if  t < 10 or t > 150: return 0
    # Loop through all activation times
    for ti in range(10, 151):  # inclusive of 150
        if (ti - delta_t) < t < (ti + delta_t):
            modifier = float(rest['stim_exposure'])
            if modifier == 0:
                return 0
            log_modifier = np.log(modifier)
            return log_modifier * (np.abs(ti - t) / delta_t)
            

    # If t not near any activation integer
    return 0

```

```python

margin = 3
start = 10
end = 150
scaling_ = {
  "weak":0.3,
  "medium":0.7,
  "strong":1,
}
def sustained_light_fn(t, rest):
    scaling = scaling_[rest['group']]
    if t < start or t > end: return 0
    elif t < start+margin: 
      return (scaling/margin)*(t-start)
    elif t > end-margin: 
      return (scaling/margin)*(end-t)
    return scaling


```

```python
times = np.linspace(1, 175, 350)
ys = np.array([sustained_light_fn(p,{"group":'weak'}) for p in times]) 
#ys = ys / max(ys)

fig, ax = plt.subplots(figsize=(9, 5))
ax.plot(times, ys, alpha=0.6)
ax.set_title('light function')
ax.set_xlabel('Time (s)')
ax.set_ylabel('light_fn value')
```
